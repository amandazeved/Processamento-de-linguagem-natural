# Contagem de Palavras em Livros - LAB 01

Objetivo: contar palavras em 3 livros diferentes.

Livros escolhidos:

* Helena - Machado de Assis https://www.gutenberg.org/ebooks/67162
* Dom Casmurro - Machado de Assis https://www.gutenberg.org/ebooks/55752
* Iracema - José de Alencar https://www.gutenberg.org/ebooks/67740

O código deste notebook é baseado em: https://www.datacamp.com/projects/38

```python
import requests
from bs4 import BeautifulSoup
import nltk
```

Pegando os livros em *r1, r2, r3*:

```python
r1 = requests.get("https://www.gutenberg.org/cache/epub/67162/pg67162-images.html")
r2 = requests.get("https://www.gutenberg.org/cache/epub/55752/pg55752-images.html")
r3 = requests.get("https://www.gutenberg.org/cache/epub/67740/pg67740-images.html")

r1.encoding = 'utf-8'

html1 = r1.text
html2 = r2.text
html3 = r3.text

print(html1[:2000])
```

Pegando só a parte textual do html de cada livro:

```python
soup1 = BeautifulSoup(html1, 'html.parser')
soup2 = BeautifulSoup(html2, 'html.parser')
soup3 = BeautifulSoup(html3, 'html.parser')

text1 = soup1.get_text()
text2 = soup2.get_text()
text3 = soup3.get_text()
```

Mostrando uma parte do texto do primeiro livro (*Helena*):

```python
print(text1[1550:3000])
```

Parte do segundo texto (*Dom Casmurro*):

```python
print(text2[1060:3000])
```

Parte do terceiro texto (*Iracema*):

```python
print(text3[14915:16000])
```

## Usando Tokenização:

```python
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')

tokens1 = tokenizer.tokenize(text1)
tokens2 = tokenizer.tokenize(text2)
tokens3 = tokenizer.tokenize(text3)
```

Mostrando os tokens do primeiro livro:

```python
print(tokens1[901:925])
print(tokens2[901:925])
print(tokens3[933:952])
```

Normalizando todos os tokens para minusculo:

```python
words1 = []

for word in tokens1:
  words1.append(word.lower())

print(words1[901:925])
```

```python
words2 = []

for word in tokens2:
  words2.append(word.lower())

print(words2[901:925])
```

```python
words3 = []

for word in tokens3:
  words3.append(word.lower())

print(words3[933:952])
```

## Fazer a filtragem de stopwords:

Foi necessário fazer a filtragem de stopwords em inglês por causa da existência dessas palavras no texto.

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
sw = stopwords.words('portuguese')
sw_en = stopwords.words('english')
```

Como as obras são antigas achei necessário acrescentar a lista de stopwords algumas palavras do português que não são mais usadas e mais algumas palavras.

```python
sw_add = ['elle', 'elles', 'ella', 'ellas', 'á', 'ás','ja','ha', 'ésta', 'tão', 'tambem', 'assim', 'ainda']
for w in sw_add:
  sw.append(w)
```

Alguns stopwords do Portugues:

```python
print(sw)
```

### Filtragem no texto 1 - Helena

```python
words_ns1 =  [word for word in words1 if (word not in sw) and (word not in sw_en)]

print("Sem a filtragem: ")
print(words1[:20])
print(words1[249:274])

print("\nCom a filtragem de stopwords")
print(words_ns1[:20])
print(words_ns1[180:195])
```

### Filtragem no texto 2 - Dom Casmurro

```python
words_ns2 =  [word for word in words2 if (word not in sw) and (word not in sw_en)]

print("Sem a filtragem: ")
print(words2[:20])
print(words2[197:220])

print("\nCom a filtragem de stopwords")
print(words_ns2[:20])
print(words_ns2[125:138])
```

### Filtragem no texto 3 - Iracema

```python
words_ns3 =  [word for word in words3 if (word not in sw) and (word not in sw_en)]

print("Sem a filtragem: ")
print(words3[:20])
print(words3[286:310])

print("\nCom a filtragem de stopwords")
print(words_ns3[:20])
print(words_ns3[208:223])
```

## Gerar tabela de frequência:

### Tabela de frequência texto 1 - Helena

```python
freqdist1 = nltk.FreqDist(words_ns1)

freqdist1
```

```python
%matplotlib inline

freqdist1.plot(30)
```

Podemos ver que as três primeira palavras mais frequentes são os nomes dos personagens do livro. *Helena* o personagem central, *Estácio* o protagonista masculino, e *Úrsula* tia de Estácio. Além de poder ver outros personagem nesse top 30, como: *Padre Melchior*, *Mendonça*, *Camargo*, *Eugênia*.

### Tabela de frequência texto 2 - Dom Casmurro

```python
freqdist2 = nltk.FreqDist(words_ns2)

freqdist2
```

```python
freqdist2.plot(30)
```

Podemos observar que o nome do protagonista *Bento Santiago* não aparece no top 30, pois a historia é contada em primeira pessoa. E a palavras mais frequente é *Capitu*, o amor do protagonista.

### Tabela de frequência texto 3 - Iracema

```python
freqdist3 = nltk.FreqDist(words_ns3)

freqdist3
```

```python
freqdist3.plot(30)
```

*Iracema* como personagem central aparece como a palavra mais frequente, depois vem guerreiro como Martim, o amor de Iracema, era muitas vezes referenciado.

Analisando os três textos podemos ver que aparece as palavras *project* e *gutemberg*, apenas mostrando que os livros foram tirados do site do Projeto Gutemberg e poderia ser tirados na filtragem já não faz parte da história.

## N-Gramas

Análise dos bigramas.

```python
from nltk import bigrams
string_bigrams1 = list(bigrams(words_ns1))
print(string_bigrams1[:10])
```

### Texto 1 - Helena

```python
fd1 = nltk.FreqDist(string_bigrams1)

fd1.plot(30)
```

Como já foi dito poderiamos tirar as palavras *Project Gutenberg* já que não agrega valor ao texto. Mas podemos observar algumas combinações interessante como *Padre Melchior* e *Dr. Camargo* que são personagem do enredo.

### Texto 2 - Dom Casmurro

```python
string_bigrams2 = list(bigrams(words_ns2))
print(string_bigrams2[:10])
```

```python
fd2 = nltk.FreqDist(string_bigrams2)

fd2.plot(30)
```

Podemos ver mais uma vez os nomes dos personagem.

### Texto 3 - Iracema

```python
string_bigrams3 = list(bigrams(words_ns3))
print(string_bigrams3[:10])
```

```python
fd3 = nltk.FreqDist(string_bigrams3)

fd3.plot(30)
```

Aqui o *guerreiro branco* faz referencia ao Martim, o protagonista masculino, e *filha de araken* (ou *filha de Araquém*) representa Iracema.
